```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load GPT-2
model_name = "gpt2"  # You can also try "gpt2-medium"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Move model to GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
```

```python
# Your prompt
prompt = "Once upon a time in a quiet village"

# Encode the prompt
input_ids = tokenizer.encode(prompt, return_tensors="pt").to(device)

# Generate text
output = model.generate(
    input_ids,
    max_length=250,
    temperature=1.0,
    top_p=0.95,
    do_sample=True,
    pad_token_id=tokenizer.eos_token_id
)

# Decode and print
story = tokenizer.decode(output[0], skip_special_tokens=True)
print("\n✅ Generated Story:\n", story)
```

```python
!pip install  nltk
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, AutoModelForSeq2SeqLM
import nltk
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from nltk.translate.rouge_score import rouge_n
import pandas as pd
```

```python
nltk.download('punkt')
```

```python
# Model metadata
MODELS = {
    "GPT-2": {
        "model_name": "gpt2",
        "causal": True
    },
    "FLAN-T5": {
        "model_name": "google/flan-t5-base",
        "causal": False
    },
    "DistilGPT-2": {
        "model_name": "distilgpt2",
        "causal": True
    }
}
```

```python
def load_model(model_info):
    tokenizer = AutoTokenizer.from_pretrained(model_info["model_name"])
    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token
    if model_info["causal"]:
        model = AutoModelForCausalLM.from_pretrained(model_info["model_name"])
    else:
        model = AutoModelForSeq2SeqLM.from_pretrained(model_info["model_name"])
    return model, tokenizer
```

```python
def preprocess_prompt(prompt, model_type):
    prompt = prompt.strip()
    if model_type in ["GPT-2", "DistilGPT-2"]:
        return f"Once upon a time, {prompt}"
    elif model_type == "FLAN-T5":
        return f"Write a story about: {prompt}"
    return prompt
```

```python
def generate_story(model_name, prompt, max_length, temperature, top_p):
    model_info = MODELS[model_name]
    model, tokenizer = load_model(model_info)
    formatted_prompt = preprocess_prompt(prompt, model_name)

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = model.to(device)

    inputs = tokenizer(formatted_prompt, return_tensors="pt", truncation=True, padding=True).to(device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_length=max_length,
            do_sample=True,
            temperature=temperature,
            top_p=top_p,
            pad_token_id=tokenizer.pad_token_id
        )

    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return generated_text.replace(formatted_prompt, "").strip()[:500] if model_info["causal"] else generated_text[:500]
```

```python
def calculate_bleu(reference, candidate):
    smoothie = SmoothingFunction().method4
    reference_tokens = [nltk.word_tokenize(reference)]
    candidate_tokens = nltk.word_tokenize(candidate)
    return sentence_bleu(reference_tokens, candidate_tokens, smoothing_function=smoothie)

def calculate_rouge(reference, candidate, n=1):
    reference_tokens = nltk.word_tokenize(reference)
    candidate_tokens = nltk.word_tokenize(candidate)
    return rouge_n([reference_tokens], [candidate_tokens], n)[0]

def evaluate_story(reference, generated):
    if not generated.strip() or not reference.strip():
        return {"bleu": 0.0, "rouge-1": 0.0, "rouge-2": 0.0}

    try:
        bleu_score = calculate_bleu(reference, generated)
        rouge1_score = calculate_rouge(reference, generated, 1)
        rouge2_score = calculate_rouge(reference, generated, 2)
    except:
        bleu_score = 0.0
        rouge1_score = 0.0
        rouge2_score = 0.0

    return {
        "bleu": bleu_score,
        "rouge-1": rouge1_score,
        "rouge-2": rouge2_score
    }
```

```python

def main():
    # Configuration
    prompt = "A lone astronaut discovers a hidden planet."
    max_length = 300
    temperature = 0.7
    top_p = 0.9
    reference = "An astronaut finds a mysterious new world with strange life forms."

    print("\n" + "="*50)
    print(f"Generating stories with prompt: '{prompt}'")
    print("="*50 + "\n")

    results = []

    for model_name in MODELS:
        print(f"\n=== Generating with {model_name} ===")
        try:
            story = generate_story(model_name, prompt, max_length, temperature, top_p)
            metrics = evaluate_story(reference, story)

            print("\nGenerated Story:")
            print("-"*40)
            print(story)
            print("-"*40)

            print(f"\nBLEU Score: {metrics['bleu']:.4f}")
            print(f"ROUGE-1: {metrics['rouge-1']:.4f}")
            print(f"ROUGE-2: {metrics['rouge-2']:.4f}")

            results.append({
                "Model": model_name,
                "BLEU": metrics["bleu"],
                "ROUGE-1": metrics["rouge-1"],
                "ROUGE-2": metrics["rouge-2"],
                "Story": story
            })

        except Exception as e:
            print(f"Error with {model_name}: {str(e)}")

    # Show comparison
    if results:
        print("\n\n=== COMPARISON RESULTS ===")
        df = pd.DataFrame(results)
        print(df[["Model", "BLEU", "ROUGE-1", "ROUGE-2"]].to_markdown(index=False))
        print("\nFull stories available in the results dataframe")
```

```python
if __name__ == "__main__":
    main()
```
